```python
#importamos librerias necesarias
import pandas as pd
```

```python
#Cargamos los datos del archivo CSV de Códigos de Muerte
df_CodigosDeMuerte = pd.read_csv(
    'ArchivosOriginales/Codigos de Muerte.csv',
    sep=';',          
    engine='python'   
)

df_CodigosDeMuerte.head() #mostramos las primeras filas del dataframe
```

```python
#Cargamos los datos del archivo CSV de Datos de mortalidad
df_DatosMortalidad = pd.read_csv(
    "ArchivosOriginales/Datos de Mortalidad.csv",
    sep=';',
    engine='python')

df_DatosMortalidad.head()#imprimimos los primeros 5 registros del dataframe
```

```python
#Cargamos los datos del archivo CSV de Datos de Divipola

df_Divipola = pd.read_csv(
    "ArchivosOriginales\Divipola.csv",
    sep =";",
    engine='python'
)

df_Divipola.head() #imprimimos los primeros 5 registros del dataframe
```

```python
#Generamos el archivo de muester por departamento en colombia para el año 2019
#agrupamos los datos del archivo datos de mortalidad por departamento 

df_MuertesPorDepartamento = df_DatosMortalidad.groupby(['COD_DEPARTAMENTO','SEXO','MANERA_MUERTE']).size().reset_index(name='Muertes') #agrupamos los datos por departamento y contamos el número de muertes
#Concatenamos el dataframe de muertes por departamento con el dataframe de divipola para obtener el nombre del departamento, Sexo, edad y manera de muerte
df_MuertesPorDepartamento = pd.merge(df_MuertesPorDepartamento, df_Divipola[['COD_DEPARTAMENTO', 'DEPARTAMENTO']], on='COD_DEPARTAMENTO', how='left') #unimos los dataframes por la columna COD_DEPARTAMENTO
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo
df_MuertesPorDepartamento['SEXO'] = df_MuertesPorDepartamento['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#Generamos el archivo CSV de muertes por departamento
ruta_Salida = "ArchivosProcesados/MuertesPorDepartamento.csv" #definimos la ruta de salida
df_MuertesPorDepartamento.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```

```python
#Generamos el archivo que muestre muertes por mes en colombia

df_MuertesPorMes = df_DatosMortalidad.groupby(['MES','SEXO','HORA','MANERA_MUERTE']).size().reset_index(name='Muertes') #agrupamos los datos por mes y contamos el número de muertes
#Añadimos el nombre del mes al dataframe
df_MuertesPorMes['MES'] = df_MuertesPorMes['MES'].replace({1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'}) #reemplazamos los valores de la columna MES
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo
df_MuertesPorMes['SEXO'] = df_MuertesPorMes['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#Generamos el archivo CSV de muertes por mes
ruta_Salida = "ArchivosProcesados/MuertesPorMes.csv" #definimos la ruta de salida
df_MuertesPorMes.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```

```python
#agrupamos los datos de mortalidad por codigo de muerte, sexo y mes 
df_MuertesPorCodigo = df_DatosMortalidad.groupby(['COD_MUERTE','SEXO','MES',]).size().reset_index(name='Muertes') #agrupamos los datos por codigo de muerte y contamos el número de muertes
#Concatenamos el dataframe de muertes por codigo de muerte con el dataframe de codigos de muerte para obtener el nombre del codigo de muerte
df_MuertesPorCodigo = pd.merge(
	df_MuertesPorCodigo,
	df_CodigosDeMuerte[['Código de la CIE-10 cuatro caracteres', "Descripcion  de códigos mortalidad a cuatro caracteres"]],
	left_on='COD_MUERTE',
	right_on='Código de la CIE-10 cuatro caracteres',
	how='left'
) #unimos los dataframes por la columna COD_MUERTE
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo 
df_MuertesPorCodigo['SEXO'] = df_MuertesPorCodigo['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#Generamos el archivo CSV de muertes por codigo de muerte
ruta_Salida = "ArchivosProcesados/MuertesPorCodigo.csv" #definimos la ruta de salida
df_MuertesPorCodigo.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```

```python
#Agrupamos los datos por municipio, Mes, Manera de muerte y contamos el número de muertes
df_MuertesPorMunicipioTabla = df_DatosMortalidad.groupby(['COD_MUNICIPIO','MES', 'SEXO', "COD_MUERTE"]).size().reset_index(name='Muertes') #agrupamos los datos por municipio y contamos el número de muertes
#Concatenamos el dataframe de muertes por municipio con el dataframe de divipola para obtener el nombre del municipio
df_MuertesPorMunicipioTabla = pd.merge(df_MuertesPorMunicipioTabla, df_Divipola[['COD_MUNICIPIO', 'MUNICIPIO']], on='COD_MUNICIPIO', how='left') #unimos los dataframes por la columna COD_MUNICIPIO  
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo
df_MuertesPorMunicipioTabla['SEXO'] = df_MuertesPorMunicipioTabla['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#unimos el dataframe con el dataframe de codigos de muerte para obtener el nombre del codigo de muerte
df_MuertesPorMunicipioTabla = pd.merge(
    df_MuertesPorMunicipioTabla,
    df_CodigosDeMuerte[['Código de la CIE-10 cuatro caracteres', "Descripcion  de códigos mortalidad a cuatro caracteres"]],
    left_on='COD_MUERTE',
    right_on='Código de la CIE-10 cuatro caracteres',
    how='left'
) #unimos los dataframes por la columna COD_MUERTE
#Generamos el archivo CSV de muertes por municipio
ruta_Salida = "ArchivosProcesados/MuertesPorMunicipioTabla.csv" #definimos la ruta de salida
#df_MuertesPorMunicipioTabla.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```

```python
df_histograma = df_DatosMortalidad.groupby(['SEXO','GRUPO_EDAD1','MES']).size().reset_index(name='Muertes')    #agrupamos los datos por edad y contamos el número de muertes
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo
df_histograma['SEXO'] = df_histograma['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#Generamos el archivo CSV de muertes por edad
ruta_Salida = "ArchivosProcesados/MuertesPorEdad.csv" #definimos la ruta de salida
df_histograma.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```

```python
#Filtramos el archivo de datos de mortalidad para obtener solo los registros de muertes por homicidio , filtramos por manera de muerte
df_DatosMortalidad = df_DatosMortalidad[df_DatosMortalidad['MANERA_MUERTE']=="Homicidio"] #filtramos los datos por manera de muerte
#Agrupamos los datos por municipio, Mes, Manera de muerte y contamos el número de muertes
df_MuertesPorMunicipio = df_DatosMortalidad.groupby(['COD_MUNICIPIO','MES', 'SEXO']).size().reset_index(name='Muertes') #agrupamos los datos por municipio y contamos el número de muertes
#Concatenamos el dataframe de muertes por municipio con el dataframe de divipola para obtener el nombre del municipio
df_MuertesPorMunicipio = pd.merge(df_MuertesPorMunicipio, df_Divipola[['COD_MUNICIPIO', 'MUNICIPIO']], on='COD_MUNICIPIO', how='left') #unimos los dataframes por la columna COD_MUNICIPIO  
#reemplazar 1 por Masculino y 2 por Femenino en la columna Sexo
df_MuertesPorMunicipio['SEXO'] = df_MuertesPorMunicipio['SEXO'].replace({1: 'Masculino', 2: 'Femenino', 3: 'No Definido'}) #reemplazamos los valores de la columna Sexo
#Convertimos la columna MES a tipo int
df_MuertesPorMunicipio['MES'] = df_MuertesPorMunicipio['MES'].astype(int) #convertimos la columna MES a tipo int
#Generamos el archivo CSV de muertes por municipio
ruta_Salida = "ArchivosProcesados/MuertesPorMunicipio.csv" #definimos la ruta de salida
df_MuertesPorMunicipio.to_csv(ruta_Salida, sep=';', index=False) #guardamos el dataframe en un archivo CSV
```
